# Libraries

```python
import pandas as pd
```

<b>Aim</b>: This script creates two dataframes with the external URLs frequencies (all URLs and in references) by pages.

# Page URLs

```python
df_urls = pd.read_csv('data/page_url.tsv', sep='\t')
df_urls
```

General external URLs frequencies are calculated.

```python
df3 = df_urls[['page_id']].groupby(['page_id']).size().reset_index(name='urls')
df3
```

General external URLs are exported.

```python
df3.to_csv('data/page_exturls_freq.tsv', sep='\t', index=False)
```

In references external URLs frequencies are calculated.

```python
df4 = df_urls[df_urls['in_reference']==1][['page_id']].groupby(['page_id']).size().reset_index(name='urls')
df4
```

In references external URLs frequencies are exported.

```python
df4.to_csv('data/page_refurls_freq.tsv', sep='\t', index=False)
```
